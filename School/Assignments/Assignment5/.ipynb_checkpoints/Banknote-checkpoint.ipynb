#### 2. for each class and for each feature f1; f2; f3; f4, compute its mean and standard deviation. Round the results to 2 decimal places and summarize them in a table as shown below?

```python
def summ(d_frame):
    li = list()
    for i in range(4):
        li.append(d_frame.mean()[i]) 
        li.append(d_frame.std()[i]) 
    df_ret = pd.DataFrame(li, index = ["mean f1","std f1","mean f2","std f2","mean f3","std f3","mean f4","std f4"]).T
    return df_ret
```

```python
df_summ = pd.concat([summ(df_banknoteC[df_banknoteC["class"] == 0]), summ(df_banknoteC[df_banknoteC["class"] == 1]), \
                     summ(df_banknoteC)]).reset_index(drop="true")

df_summ.insert(0,column="class",value=[0,1,"all"])
df_summ.round(2)
```

## Question 3

#### 1. take k = 3; 5; 7; 9; 11. For each k, generate Xtrain and Xtest using 50/50 split as before. Train your k-NN classifier on Xtrain and compute its accuracy for Xtest

```python
def knn_class(k):
    x_train_knn, x_test_knn, y_train_knn, y_test_knn  = \
    train_test_split(df_banknoteC.iloc[:,:4], df_banknoteC["class"], test_size = 0.5,random_state=2)
    x = x_train_knn.values
    y = y_train_knn.values
    scaler = StandardScaler().fit(x)
    x = scaler.transform(x)
    neigh = KNeighborsClassifier(n_neighbors = k)
    neigh.fit(x,y)
    new_scaled_x_test = scaler.transform(x_test_knn)
    y_predict = neigh.predict(new_scaled_x_test)
    return [y_test_knn.values,y_predict]
```

* y_test = Actual and Prediction values for each K                             
* y_con_mat = tn, fp, fn, tp for each K
* y_acc = Accuracy for each K

```python
y_test = dict()                              
y_con_mat = dict()
y_acc = dict()
for i in [3,5,7,9,11]:
    y_test[str(i)] = knn_class(i) 
    y_con_mat[str(i)] = confusion_matrix(y_test[str(i)][0], y_test[str(i)][1])
    y_acc[str(i)] = (y_con_mat[str(i)][1][1] + y_con_mat[str(i)][0][0]) / np.sum(y_con_mat[str(i)])
```

```python
df_k_acc = pd.DataFrame(y_acc, index = [0])
df_k_acc
```

#### 2. plot a graph showing the accuracy. On x axis you plot k and on y-axis you plot accuracy. What is the optimal value k* of k?

```python
plt.plot(list(y_acc.keys()),list(y_acc.values()))
plt.xlabel("K",fontsize = 14)
plt.ylabel("Accuracy",fontsize = 14)
plt.title("K - Accuracy",fontsize = 16)
plt.show()
```

#### 3. use the optimal value k* to compute performance measures and summarize them in the table

```python
tp7 = y_con_mat["7"][1][1]
tn7 = y_con_mat["7"][0][0]
fp7 = y_con_mat["7"][0][1]
fn7 = y_con_mat["7"][1][0]
accuracy7 = round(y_acc["7"],4)
tpr7 = round(tp7 / (tp7 + fn7),4)
tnr7 = round(tn7 / (tn7 + fp7),4)
```

```python
df_k7 = pd.DataFrame([[tp7,fp7,tn7,fn7,accuracy7,tpr7,tnr7]], columns = ["TP","FP","TN","FN","Accuracy","TPR","TNR"])
df_k7
```

## Question 4

#### 1. take your best value k*. For each of the four features f1; : : : ; f4, generate new Xtest and Xtrain and drop that feature from both Xtrain and Xtest. Train your classifier on the "truncated" Xtrain and predict labels on Xtest using just 3 remaining features. You will repeat this for 4 cases: (1) just f1 is missing, (2) just f2 missing, (3) just f3 missing and (4) just f4 is missing. Compute the accuracy for each of these scenarious.

```python
def feature_selection(d_frame,d_frame1):
    x_train_knn_fs, x_test_knn_fs, y_train_knn_fs, y_test_knn_fs  = \
    train_test_split(d_frame, d_frame1, test_size = 0.5,random_state=2)
    x_fs = x_train_knn_fs.values
    y_fs = y_train_knn_fs.values
    
    scaler_fs = StandardScaler().fit(x_fs)
    x_fs = scaler_fs.transform(x_fs)
    
    neigh_fs = KNeighborsClassifier(n_neighbors = 7)
    neigh_fs.fit(x_fs,y_fs)
    
    new_scaled_x_test_fs = scaler_fs.transform(x_test_knn_fs)
    y_predict_fs = neigh_fs.predict(new_scaled_x_test_fs)
    
    return [y_test_knn_fs.values,y_predict_fs]
```

```python
wo_v = feature_selection(df_banknoteC[["skewness","curtosis","entropy"]], df_banknoteC["class"])
wo_s = feature_selection(df_banknoteC[["variance","curtosis","entropy"]], df_banknoteC["class"])
wo_c = feature_selection(df_banknoteC[["variance","skewness","entropy"]], df_banknoteC["class"])
wo_e = feature_selection(df_banknoteC[["variance","skewness","curtosis"]], df_banknoteC["class"])
```

```python
tn_wo_v, fp_wo_v, fn_wo_v, tp_wo_v =  confusion_matrix(wo_v[0], wo_v[1]).ravel()
tn_wo_s, fp_wo_s, fn_wo_s, tp_wo_s =  confusion_matrix(wo_s[0], wo_s[1]).ravel()
tn_wo_c, fp_wo_c, fn_wo_c, tp_wo_c =  confusion_matrix(wo_c[0], wo_c[1]).ravel()
tn_wo_e, fp_wo_e, fn_wo_e, tp_wo_e =  confusion_matrix(wo_e[0], wo_e[1]).ravel()
```

```python
accuracy_wo_v = (tp_wo_v + tn_wo_v) / len(wo_v[0])
accuracy_wo_s = (tp_wo_s + tn_wo_s) / len(wo_s[0])
accuracy_wo_c = (tp_wo_c + tn_wo_c) / len(wo_c[0])
accuracy_wo_e = (tp_wo_e + tn_wo_e) / len(wo_e[0])
```

```python
print(f"""
Accuracy without variance {accuracy_wo_v:.4f} and
Accuracy without skewness {accuracy_wo_s:.4f} and
Accuracy without curtosis {accuracy_wo_c:.4f} and
Accuracy without entropy {accuracy_wo_e:.4f}
""")
```

## Question 5 Logistic Regression

#### 1. Use 50/50 split to generate new Xtrain and Xtest. Train your logistic regression classifier on Xtrain and compute its accuracy for Xtest

```python
def lr(d_frame,d_frame1):
    x_train_lr, x_test_lr, y_train_lr, y_test_lr  = \
    train_test_split(d_frame, d_frame1, test_size = 0.5,random_state=2)
    
    x_lr = x_train_lr.values
    y_lr = y_train_lr.values
    
    scaler_lr = StandardScaler().fit(x_lr)
    x_lr = scaler_lr.transform(x_lr)
    
    log_reg_clss = LogisticRegression()
    log_reg_clss.fit(x_lr,y_lr)
    
    new_scaled_x_lr = scaler_lr.transform(np.asmatrix(x_test_lr))
    p_lr = log_reg_clss.predict(new_scaled_x_lr)
    acc_lr = log_reg_clss.score(x_lr, y_lr)
    
    return [y_test_lr.values,p_lr,acc_lr]    
```

```python
a_lr = lr(df_banknoteC[["variance","skewness","curtosis","entropy"]], df_banknoteC["class"])
tn_lr, fp_lr, fn_lr, tp_lr = confusion_matrix(a_lr[0], a_lr[1]).ravel()
tnr_lr =  round(tn_lr / (tn_lr + fp_lr),4)
tpr_lr = round(tp_lr / (tp_lr + fn_lr),4)
```

#### 2. summarize your performance measures in the table

```python
df_lr = pd.DataFrame([[tp_lr,fp_lr,tn_lr,fn_lr,a_lr[2],tpr_lr,tnr_lr]], columns = ["TP","FP","TN","FN","Accuracy","TPR","TNR"])
df_lr
```
